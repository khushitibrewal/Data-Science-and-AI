# Module 4: Point Estimation 🎯

A **point estimate** is our single best guess for an unknown population parameter. For example, if we calculate the mean of a sample, that sample mean is a point estimate for the true population mean.

In this module, we'll discuss how to find point estimates and what makes a 'good' estimator.

## 4.1 What is a Point Estimate?

Let's take a sample of data and calculate some point estimates for its corresponding population parameters.

- The **sample mean** ($\bar{x}$) is a point estimate for the **population mean** ($\mu$).
- The **sample proportion** ($\hat{p}$) is a point estimate for the **population proportion** ($p$).
- The **sample variance** ($s^2$) is a point estimate for the **population variance** ($\sigma^2$).

```python
import numpy as np
import pandas as pd

# Let's generate a sample of exam scores for 50 students.
# Assume the scores are out of 100.
np.random.seed(123)
scores_sample = np.random.randint(50, 100, size=50)

print("Sample of 50 student scores:")
print(scores_sample)

# --- Calculating Point Estimates --- #

# 1. Point estimate for the population mean score (μ)
sample_mean = np.mean(scores_sample)
print(f"\nPoint Estimate for the Population Mean Score (μ): {sample_mean:.2f}")

# 2. Point estimate for the population proportion of students who passed (p)
# Let's assume the passing score is 65.
passing_students = np.sum(scores_sample >= 65)
total_students = len(scores_sample)
sample_proportion = passing_students / total_students
print(f"Point Estimate for the Population Proportion of Passing Students (p): {sample_proportion:.2f}")

# 3. Point estimate for the population variance of scores (σ^2)
# Note: When estimating population variance from a sample, we use 'n-1' in the denominator.
# This is called Bessel's correction and it provides an unbiased estimate.
sample_variance = np.var(scores_sample, ddof=1) # ddof=1 means degrees of freedom = n-1
print(f"Point Estimate for the Population Variance of Scores (σ^2): {sample_variance:.2f}")
```

## 4.2 Properties of Good Estimators

How do we know if our estimator is any good? We evaluate estimators based on a few key properties. Let's focus on the most important one: **Unbiasedness**.

### Unbiased Estimator

An estimator is **unbiased** if the mean of its sampling distribution is equal to the true value of the parameter being estimated. In other words, on average, the estimator hits the true target. The sample mean ($\bar{x}$) is an unbiased estimator of the population mean ($\mu$).

Let's demonstrate why we use `n-1` when calculating sample variance. We'll show that using `n` gives a *biased* estimate, while `n-1` gives an *unbiased* estimate.

```python
# Let's create a population with a known variance.
np.random.seed(42)
population = np.random.normal(loc=100, scale=15, size=10000)
true_population_variance = np.var(population)
print(f"True Population Variance (σ^2): {true_population_variance:.2f}")

# Now, let's draw many samples and calculate variance both ways.
num_samples = 5000
sample_size = 10
biased_variances = []   # Using denominator 'n'
unbiased_variances = [] # Using denominator 'n-1'

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    # Biased calculation
    biased_variances.append(np.var(sample, ddof=0))
    # Unbiased calculation
    unbiased_variances.append(np.var(sample, ddof=1))

# Now let's find the average of our estimated variances.
# If the estimator is unbiased, its average should be very close to the true population variance.
avg_biased_var = np.mean(biased_variances)
avg_unbiased_var = np.mean(unbiased_variances)

print(f"\nAverage of BIASED sample variances (using n): {avg_biased_var:.2f}")
print(f"Average of UNBIASED sample variances (using n-1): {avg_unbiased_var:.2f}")

# --- Conclusion --- #
# Notice that the average of the biased variances is consistently lower than the true variance.
# The average of the unbiased variances (using n-1) is almost exactly equal to the true variance.
# This is why we use n-1! It corrects the systematic underestimation.
```

### Other Properties (Briefly)

- **Efficiency**: An efficient estimator has a smaller variance in its sampling distribution. Given two unbiased estimators, we prefer the more efficient one because it's more likely to be close to the true value.
- **Consistency**: A consistent estimator gets closer to the true parameter value as the sample size increases.

### Key Takeaway 💡

A point estimate is our single best numerical guess for a population parameter based on sample data. We prefer estimators that are **unbiased** (correct on average) and **efficient** (low variability).
